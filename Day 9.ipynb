```python
import pyttsx3

def convert_text_to_speech(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

def main():
    text_list = ["shoutout to this.", "This is the second text.", "This is the third text."]
    for text in text_list:
        convert_text_to_speech(text)

if __name__ == "__main__":
    main()
```
